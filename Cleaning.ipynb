```python
# defining a function for making copy of data.

def data(df_train, df_test):
    df1, df2 = df_train.copy(), df_test.copy()
    df1["group"] = "df_train"
    df2["group"] = "df_test"

    return pd.concat([df1, df2], axis = 0, ignore_index = True)

data = data(df_train, df_test)
data.head(5)
```

```python
# defining a function that takes two parameters: a dataframe and an ordinal threshold.

def ColDataTypes(data, ord):
    numerical = data.select_dtypes("number").columns.to_list()
    categorical = data.select_dtypes("object").columns.to_list()

    # identifies columns with a unique value count less than the 
    # specified threshold (ord) from the numeric columns. 
    # These are considered as potential ordinal columns.
    ordinals = [col for col in numerical if data[col].nunique() < ord]

    # Updating the list of numeric columns by excluding the identified ordinal columns.
    numerical = [col for col in numerical if col not in ordinals]
    return numerical, ordinals, categorical

numerical, ordinals, categorical = ColDataTypes(df_test, 20)

print(f"Columns list for Numericals: {numerical}", end = "\n\n")
print(f"Columns list for Categoricals: {categorical}", end = "\n\n")
print(f"Columns list for Ordinals: {ordinals}", end = "\n\n")
```
